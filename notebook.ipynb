```python
#import recipes

import json

with open("recipes.json", "r") as f:
              recipes = json.load(f)
recipes
```

```python
from collections import defaultdict

def generate_shopping_list(selected_recipes, recipes):
    shopping = defaultdict(lambda: {"qty": 0, "unit": ""})

    for recipe in selected_recipes:
        for ingredient, details in recipes[recipe].items():
            
            if shopping[ingredient]["unit"] and shopping[ingredient]["unit"] != details["unit"]:
                raise ValueError(f"Unit mismatch for {ingredient}")

            shopping[ingredient]["qty"] += details["qty"]
            shopping[ingredient]["unit"] = details["unit"]

    return shopping
```

```python
print("Available recipes:")
for i, r in enumerate(recipes.keys()):
    print(f"{i}: {r}")

choice = input("Enter recipe numbers (comma separated): ")

selected = [list(recipes.keys())[int(i.strip())] for i in choice.split(",")]

shopping_list = generate_shopping_list(selected, recipes)

print("\nShopping List:")
for item, d in shopping_list.items():
    print(f"{item}: {d['qty']} {d['unit']}")
```
